**Data** **Cleaning**

Remove duplicate rows

```python
print("Duplicates before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())
```

**Convert DOB to datetime and create AGE**

```python
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
df['AGE'] = df['DOB'].apply(lambda dob: datetime.now().year - dob.year if pd.notnull(dob) else np.nan)
```

**Drop rows where DOB could not be parsed ( this is optional)**

```python
df = df[df['DOB'].notnull()]
```

**Handle DAYTRIPSBYWEIGHT:**

Convert to numeric and drop placeholder values (e.g. 999999.999999)

```python
if 'daytripsBYweight' in df.columns:

    df['daytripsBYweight'] = pd.to_numeric(df['daytripsBYweight'], errors='coerce')

    df = df[df['daytripsBYweight'] < 999999]

    print("Day trips data cleaned successfully.")
    print(f"Remaining records: {len(df)}")
else:
    print("Column 'daytripsBYweight' not found. Available columns:")
    print([col for col in df.columns if 'trip' in col.lower()])
```

Clean up PROV (Province), SEX, and POPULATION

```python
df = pd.read_csv("/content/drive/MyDrive/Tp2 Project/dts-2022-day-v1.csv", encoding='ISO-8859-1')

df['POPULATION'] = df['POPULATION'].replace(['unspecified', 'Unknown', ''], pd.NA)

print(df[['PROV', 'SEX', 'POPULATION']].head())

print("\nPopulation Group Distribution:")
print(df['POPULATION'].value_counts(dropna=False))
```

Strip whitespace from text columns(0ptional)

```python
df['PROV'] = df['PROV'].str.strip()
df['SEX'] = df['SEX'].str.strip()
df['POPULATION'] = df['POPULATION'].str.strip()

print(df[['PROV', 'SEX', 'POPULATION']].head())
```

Display the cleaned dataset shape and sample

```python
df['daytripsBYweight'] = df['daytripsBYweight'].replace(999999.999999, np.nan)

df = df[df['daytripsBYweight'].notna()]

print("Cleaned dataset shape:", df.shape)
print("\nSample data:")
df[['PROV', 'SEX', 'DOB', 'POPULATION', 'daytripsBYweight']].head()
```

Outlier Dectection

Visualize with a boxplot before removing outliers

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x='daytripsBYweight')
plt.title('Boxplot of Day Trips by Weight')
plt.xlabel('Day Trips by Weight')
plt.show()
```

Use IQR to remove outliers

```python
Q1 = df['daytripsBYweight'].quantile(0.25)
Q3 = df['daytripsBYweight'].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
```

Define lower and upper bounds

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

Remove outliers

```python
print('daytripsBYweight' in df.columns)

Q1 = df['daytripsBYweight'].quantile(0.25)
Q3 = df['daytripsBYweight'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

before = df.shape[0]
df = df[(df['daytripsBYweight'] >= lower_bound) &
        (df['daytripsBYweight'] <= upper_bound)]
after = df.shape[0]

print(f"Removed {before - after} outliers using IQR method.")
print(f"Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")
```

Visualize after outlier removal

```python

plt.figure(figsize=(10, 4))

sns.boxplot(data=df, x='daytripsBYweight', color='lightblue')

plt.title('Distribution of Day Trips by Weight (After Outlier Removal)', pad=20)
plt.xlabel('Weighted Day Trips', labelpad=10)
plt.ylabel('Frequency', labelpad=10)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
